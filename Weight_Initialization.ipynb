Part 1: Understanding Weight Initialization
1. Importance of Weight Initialization

Weight initialization is a crucial step in training artificial neural networks. The weights determine how input signals are transformed as they pass through the network layers.

Careful initialization is important because:

Faster Convergence
Proper initialization helps the optimization algorithm converge faster during training.

Avoiding Vanishing or Exploding Gradients
Poor initialization can cause gradients to become extremely small or extremely large during backpropagation.

Stable Training
Balanced initial weights help the network learn stable patterns.

Breaking Symmetry
If all neurons start with identical weights, they will learn the same features.

Therefore, proper initialization ensures that each neuron learns different and meaningful features.

2. Challenges with Improper Weight Initialization

Improper initialization can create several issues:

1. Vanishing Gradient Problem

Gradients become extremely small during backpropagation.

Effects:

Early layers learn very slowly

Training becomes inefficient

2. Exploding Gradient Problem

Gradients become extremely large.

Effects:

Model becomes unstable

Training diverges

3. Slow Convergence

If weights are poorly initialized, optimization algorithms require more iterations.

4. Symmetry Problem

If weights are initialized identically, neurons produce identical outputs and fail to learn diverse features.

3. Role of Variance in Weight Initialization

Variance represents the spread of weight values.

When initializing weights, controlling the variance is important because:

Too high variance → exploding activations

Too low variance → vanishing activations

Proper initialization ensures that:

The variance of activations remains stable across layers

Information flows smoothly through the network

Initialization methods like Xavier and He initialization are specifically designed to maintain proper variance.

Part 2: Weight Initialization Techniques
1. Zero Initialization

Zero initialization means all weights are set to zero.

Example:

weights = np.zeros((3,3))

```python
weights = np.zeros((3,3))
```

Limitations

Causes symmetry problem

All neurons learn the same features

Network fails to learn complex representations

When It Is Appropriate

Zero initialization can sometimes be used for bias terms, but it is generally not recommended for weights.

2. Random Initialization

Random initialization assigns random values to weights.

Example:

```python
weights = np.random.randn(3,3) * 0.01
```

Advantages:

Breaks symmetry

Allows neurons to learn different features

However, if values are too large or too small, it can cause:

vanishing gradients

exploding gradients

To mitigate these issues:

Use small random numbers

Use scaled random initialization methods

Part 3: Applying Weight Initialization

We will train a neural network using different initialization techniques and compare performance.

Load Dataset

We will use the MNIST dataset.

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0
```

Function to Build Model

```python
def create_model(initializer):

    model = keras.Sequential([
        layers.Flatten(input_shape=(28,28)),
        layers.Dense(128, activation='relu', kernel_initializer=initializer),
        layers.Dense(64, activation='relu', kernel_initializer=initializer),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

1. Zero Initialization Model

```python
zero_init = tf.keras.initializers.Zeros()

model_zero = create_model(zero_init)

history_zero = model_zero.fit(
    X_train, y_train,
    epochs=5,
    validation_data=(X_test,y_test)
)
```

2. Random Initialization Model

```python
random_init = tf.keras.initializers.RandomNormal(mean=0,stddev=0.05)

model_random = create_model(random_init)

history_random = model_random.fit(
    X_train, y_train,
    epochs=5,
    validation_data=(X_test,y_test)
)
```

3. Xavier Initialization Model

```python
xavier_init = tf.keras.initializers.GlorotUniform()

model_xavier = create_model(xavier_init)

history_xavier = model_xavier.fit(
    X_train, y_train,
    epochs=5,
    validation_data=(X_test,y_test)
)
```

4. He Initialization Model

```python
he_init = tf.keras.initializers.HeNormal()

model_he = create_model(he_init)

history_he = model_he.fit(
    X_train, y_train,
    epochs=5,
    validation_data=(X_test,y_test)
)
```

Performance Comparison

```python
print("Zero Initialization Accuracy:",
      model_zero.evaluate(X_test,y_test)[1])

print("Random Initialization Accuracy:",
      model_random.evaluate(X_test,y_test)[1])

print("Xavier Initialization Accuracy:",
      model_xavier.evaluate(X_test,y_test)[1])

print("He Initialization Accuracy:",
      model_he.evaluate(X_test,y_test)[1])
```

| Initialization | Performance            |
| -------------- | ---------------------- |
| Zero           | Very poor learning     |
| Random         | Moderate performance   |
| Xavier         | Good convergence       |
| He             | Best for ReLU networks |


Choosing the Right Initialization Technique

When selecting an initialization method, consider:

Activation Function

Sigmoid / Tanh → Xavier initialization

ReLU → He initialization

Network Depth

Deep networks require stable variance across layers.

Model Complexity

Large models benefit from advanced initialization techniques.

Training Stability

Proper initialization improves gradient flow and convergence speed.

Conclusion

Weight initialization plays a critical role in neural network training.

Key points:

Prevents vanishing and exploding gradients

Improves convergence speed

Enables stable learning

Among the methods tested:

Zero initialization performs poorly

Random initialization works moderately

Xavier and He initialization provide better results

He initialization is most suitable for deep ReLU networks.
